```python

import pandas as pd
import numpy as np 
df_notes=pd.read_excel('/content/drive/MyDrive/Colab Notebooks/notes.xlsx')
print(df_notes)
```

```python
#print the first five lines
df_notes.head()
```

```python
ser_moy=df_notes['EXAMEN']*0.5+df_notes['TP']*0.2+df_notes['DS']*0.3

df_moy=ser_moy.to_frame(name="MOYENNE")

df_notes=pd.concat([df_notes,df_moy],axis=1)
df_notes
```

```python

#Display the CIN numbers of students with the maximum exam score underform of a dataframe, named 𝒅𝒇_𝒎𝒂𝒙
df_max=df_notes[df_notes['EXAMEN']==df_notes['EXAMEN'].max()][['CIN']]
print(df_max)
```

Add a column named AVERAGE that contains the average of each
student in the subject "Artificial Intelligence". The calculation of the average is done according to
the following coefficients:
• EXAMINATION coefficient= 0.5
• TP coefficient= 0.2
• DS coefficient= 0.3

```python
#Calculate and display the general average of the whole class in the subject"Artificial intelligence"
df_notes['MOYENNE'].mean()
```

```python
#Remove from 𝒅𝒇_𝒏𝒐𝒕𝒆𝒔 students who do not have the average in the subject "Artificial intelligence". Show 𝒅𝒇_𝒏𝒐𝒕𝒆
indexNames = df_notes[df_notes['MOYENNE'] < 10].index
df_notes.drop(index=list(indexNames), inplace=True)
df_notes
```

Ajouter une colonne, nommée MENTION, dans 𝒅𝒇_𝒏𝒐𝒕𝒆𝒔 qui comporte des valeurs nan. Puis,
remplacer selon la moyenne la valeur nan par "Passable", "Assez Bien", "Bien" ou "Trés Bien"
• Si MOYENNE < 𝟏𝟐 alors MENTION="Passable"
• Si 𝟏𝟐 ≤ MOYENNE < 𝟏𝟒 alors MENTION="Assez Bien"
• Si 𝟏𝟒 ≤ MOYENNE < 𝟏𝟔 alors MENTION="Bien"
• Si MOYENNE ≥ 𝟏𝟔 alors MENTION="Trés Bien

```python
df_notes["MENTION"] = np.nan
df_notes.loc[df_notes["MOYENNE"]< 12, "MENTION"]="Passable"
df_notes.loc[(df_notes["MOYENNE"]>= 12) & (df_notes["MOYENNE"]< 14), "MENTION"]="Assez Bien"
df_notes.loc[(df_notes["MOYENNE"]>= 14) & (df_notes["MOYENNE"]< 16) , "MENTION"]="Bien"
df_notes.loc[df_notes["MOYENNE"]>=16, "MENTION"]="Trés Bien"
df_notes
```

```python
#Trier les lignes de 𝒅𝒇_𝒏𝒐𝒕𝒆𝒔 selon l’ordre alphabétique croissant de la colonne NOM&PRENOM (réinitialiser les indices de lignes de 𝒅𝒇_𝒏𝒐𝒕𝒆𝒔). Afficher 𝒅𝒇_𝒏𝒐𝒕𝒆
df_notes.sort_values(by="NOM&PRENOM", ascending=True,
inplace=True, ignore_index=True)
df_notes
```

```python
#- Transformer 𝒅𝒇_𝒏𝒐𝒕𝒆𝒔 en deux tableaux ndarray 𝑿 et 𝒀; 𝑿 contient les valeurs de six premières colonnes, et 𝒀 contient les valeurs de la colonne MENTION. Afficher 𝑿 et y
X = df_notes.iloc[:,:df_notes.shape[1]-1].values
Y = df_notes.iloc[:, df_notes.shape[1]-1].values
print("X=\n", X)
print("Y=\n", Y)
```
