```python

import pandas as pd
import numpy as np 
df_notes=pd.read_excel('/content/drive/MyDrive/Colab Notebooks/notes.xlsx')
print(df_notes)
```

```python
ser_moy=df_notes['EXAMEN']*0.5+df_notes['TP']*0.2+df_notes['DS']*0.3

df_moy=ser_moy.to_frame(name="MOYENNE")

df_notes=pd.concat([df_notes,df_moy],axis=1)
df_notes
```

```python
#Remove from ùíÖùíá_ùíèùíêùíïùíÜùíî students who do not have the average in the subject "Artificial intelligence". Show ùíÖùíá_ùíèùíêùíïùíÜ
indexNames = df_notes[df_notes['MOYENNE'] < 10].index
df_notes.drop(index=list(indexNames), inplace=True)
df_notes
```

Ajouter une colonne, nomm√©e MENTION, dans ùíÖùíá_ùíèùíêùíïùíÜùíî qui comporte des valeurs nan. Puis,
remplacer selon la moyenne la valeur nan par "Passable", "Assez Bien", "Bien" ou "Tr√©s Bien"
‚Ä¢ Si MOYENNE < ùüèùüê alors MENTION="Passable"
‚Ä¢ Si ùüèùüê ‚â§ MOYENNE < ùüèùüí alors MENTION="Assez Bien"
‚Ä¢ Si ùüèùüí ‚â§ MOYENNE < ùüèùüî alors MENTION="Bien"
‚Ä¢ Si MOYENNE ‚â• ùüèùüî alors MENTION="Tr√©s Bien

```python
df_notes["MENTION"] = np.nan
df_notes.loc[df_notes["MOYENNE"]< 12, "MENTION"]="Passable"
df_notes.loc[(df_notes["MOYENNE"]>= 12) & (df_notes["MOYENNE"]< 14), "MENTION"]="Assez Bien"
df_notes.loc[(df_notes["MOYENNE"]>= 14) & (df_notes["MOYENNE"]< 16) , "MENTION"]="Bien"
df_notes.loc[df_notes["MOYENNE"]>=16, "MENTION"]="Tr√©s Bien"
df_notes
```

```python
#Trier les lignes de ùíÖùíá_ùíèùíêùíïùíÜùíî selon l‚Äôordre alphab√©tique croissant de la colonne NOM&PRENOM (r√©initialiser les indices de lignes de ùíÖùíá_ùíèùíêùíïùíÜùíî). Afficher ùíÖùíá_ùíèùíêùíïùíÜ
df_notes.sort_values(by="NOM&PRENOM", ascending=True,
inplace=True, ignore_index=True)
df_notes
```

```python
#- Transformer ùíÖùíá_ùíèùíêùíïùíÜùíî en deux tableaux ndarray ùëø et ùíÄ; ùëø contient les valeurs de six premi√®res colonnes, et ùíÄ contient les valeurs de la colonne MENTION. Afficher ùëø et y
X = df_notes.iloc[:,:df_notes.shape[1]-1].values
Y = df_notes.iloc[:, df_notes.shape[1]-1].values
print("X=\n", X)
print("Y=\n", Y)
```
